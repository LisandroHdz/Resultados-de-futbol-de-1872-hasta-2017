```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import read_csv
df = read_csv('/content/results 2.csv')
print(df.head())
```

# CONTEO DE PARTIDOS POR EQUIPO DE LOCAL

```python
    # Suponiendo que tus datos están en un DataFrame llamado 'df' y la columna es 'Nombre_Columna'
    conteo_valores = df['home_team'].value_counts()
    print(conteo_valores)
```

# **Top 10 equipos con mas partidos jugados de local**

---



```python
plt.figure(figsize=(15, 8)) # Define el tamaño del gráfico

# Obtener los 10 valores más frecuentes
top_10 = conteo_valores.head(10)

# Crear el gráfico de barras con los 10 valores más frecuentes
top_10.plot(kind='bar',
            color='skyblue',        # Color de barras
           edgecolor='black',      # Borde de barras
)
plt.title('Top 10 equipos con mas partidos jugados de local')
plt.xlabel('EQUIPO')
plt.ylabel('PARTIDOS')
plt.xticks(rotation=45, ha='right') # Rota las etiquetas del eje X para mejor legibilidad

# ✅ AÑADIR LOS VALORES EN LAS BARRAS
for i, (equipo, cantidad) in enumerate(top_10.items()):
    plt.text(i,                            # Posición X (índice de la barra)
             cantidad + 0.02 * max(top_10), # Posición Y (ligeramente arriba de la barra)
             str(cantidad),                # Texto a mostrar
             ha='center',                  # Alineación horizontal centrada
             va='bottom',                  # Alineación vertical en la base
             fontweight='bold',            # Texto en negrita
             fontsize=10)                  # Tamaño de fuente

plt.grid(axis='y', alpha=0.3, linestyle='--') # Líneas de grid para mejor lectura
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Top 10 equipos en ambas categorías
top_local = df['home_team'].value_counts().head(10)
top_visitante = df['away_team'].value_counts().head(10)
conteo_combinado = (df['home_team'].value_counts() + df['away_team'].value_counts()).head(10)
conteo_combinado = conteo_combinado.sort_values(ascending=False)

# Gráfico comparativo - CORREGIDO
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))  # ← 1 fila, 3 columnas

top_local.plot(kind='bar', ax=ax1, color='lightcoral', title='Top 10 Equipos como Local')
top_visitante.plot(kind='bar', ax=ax2, color='lightblue', title='Top 10 Equipos como Visitante')
conteo_combinado.plot(kind='bar', ax=ax3, color='lightgreen', title='Top 10 Equipos - Total Local + Visitante')

# Rotar etiquetas para mejor legibilidad
ax1.tick_params(axis='x', rotation=45)
ax2.tick_params(axis='x', rotation=45)
ax3.tick_params(axis='x', rotation=45)

# Añadir valores en las barras
for ax in [ax1, ax2, ax3]:
    for bar in ax.patches:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calcular los top 10 CORRECTAMENTE
top_local = df['home_team'].value_counts().nlargest(10)  # ← Más claro que head(10)
top_visitante = df['away_team'].value_counts().nlargest(10)
conteo_combinado = (df['home_team'].value_counts() + df['away_team'].value_counts()).nlargest(10)

# Gráfico
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(22, 7))

# Configurar gráficos
graphs = [
    (top_local, 'lightcoral', 'Top 10 como Local'),
    (top_visitante, 'lightblue', 'Top 10 como Visitante'),
    (conteo_combinado, 'lightgreen', 'Top 10 Total')
]

for i, (data, color, title) in enumerate(graphs):
    data.plot(kind='bar', ax=[ax1, ax2, ax3][i], color=color, edgecolor='black')
    [ax1, ax2, ax3][i].set_title(title, fontweight='bold', fontsize=12)
    [ax1, ax2, ax3][i].tick_params(axis='x', rotation=45)

    # Añadir valores
    for bar in [ax1, ax2, ax3][i].patches:
        height = bar.get_height()
        [ax1, ax2, ax3][i].text(bar.get_x() + bar.get_width()/2., height + 5,
                               f'{int(height)}', ha='center', va='bottom',
                               fontweight='bold', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# df['torneo']: Accede a la columna llamada 'torneo' en el DataFrame
# == 'FIFA World Cup': Crea una condición que busca exactamente el texto "FIFA World Cup"
# df[condición]: Filtra el DataFrame, manteniendo solo las filas donde la condición es True
df_mundial = df[df['tournament'] == 'FIFA World Cup'].copy()

print(f"Total de partidos de Mundial en el dataset: {len(df_mundial)}")
def determinar_ganador(row):
    if row['home_score'] > row['away_score']:
        return row['home_team']
    elif row['away_score'] > row['home_score']:
        return row['away_team']
    else:
        return 'Empate'

# Aplicar la función
df_mundial['ganador'] = df_mundial.apply(determinar_ganador, axis=1)
```

```python
# Contar victorias (excluyendo empates)
victorias_mundial = df_mundial[df_mundial['ganador'] != 'Empate']['ganador'].value_counts()

print("EQUIPOS CON MÁS VICTORIAS EN MUNDIALES:")
print(victorias_mundial.head(10))
```

```python
import matplotlib.pyplot as plt

top_10_victorias = victorias_mundial.head(10)

plt.figure(figsize=(12, 6))
bars = plt.bar(top_10_victorias.index, top_10_victorias.values, color='skyblue', edgecolor='black')

plt.title('Top 10 Equipos con Más Victorias en Partidos de Mundial', fontsize=14, fontweight='bold')
plt.xlabel('Equipos')
plt.ylabel('Número de Victorias')
plt.xticks(rotation=45, ha='right')

# Añadir valores en las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
